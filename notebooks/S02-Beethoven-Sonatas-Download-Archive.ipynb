## Scrape the webpage


The code is designed to automatically download and extract a series of
ZIP files containing MIDI compositions of Beethoven's piano sonatas from
a specified website.

1. **Setting Up the Environment**:
    - The code begins by importing necessary libraries: `requests` for
      downloading files from the web, `os` for directory and file
      operations, and `zipfile` for handling ZIP files.

2. **Defining URLs and Directories**:
    - The base URL `https://www.suzumidi.com/midi/beethovn/` is where
      the ZIP files are hosted.
    - A list of filenames (`file_names`) is generated using a list
      comprehension. This list contains filenames in the format
      `psXX.zip`, where `XX` ranges from `01` to `39`, representing the
      39 ZIP files to be downloaded.
    - Two directories are defined: `download_directory` for saving the
      downloaded ZIP files and `unzip_directory` for saving the
      extracted contents of these ZIP files.

3. **Preparing the Filesystem**:
    - The code checks if the directories `download_directory` and
      `unzip_directory` exist. If they don't, the code creates them.

4. **Downloading and Extracting Files**:
    - For each filename in the `file_names` list:
        - The full URL of the file is constructed by appending the
          filename to the base URL.
        - The file is then downloaded using the `requests.get()` method.
          The response is streamed to handle potentially large files
          without consuming excessive memory.
        - If the download is successful (HTTP status code 200), the file
          is saved to the `download_directory`.
        - The saved ZIP file is then extracted using the
          `zipfile.ZipFile` class, and its contents are saved to the
          `unzip_directory`.
        - A message is printed to the console indicating the progress
          (whether a file was successfully downloaded and extracted or
          if there was a failure).

5. **Completion**:
    - Once all files have been processed, a message is printed to
      indicate that the operation is complete.


```python
import requests
import os
import zipfile

# Base URL
base_url = "https://www.suzumidi.com/midi/beethovn/"

# List of file names
file_names = [f"ps{i:02}.zip" for i in range(1, 40)]

# Directory to save the downloaded zip files
download_directory = "downloaded_files"

# Directory to save the unzipped files
unzip_directory = "../data/external"

# Ensure the directories exist
if not os.path.exists(download_directory):
    os.makedirs(download_directory)
    
if not os.path.exists(unzip_directory):
    os.makedirs(unzip_directory)

# Download and unzip each file
for file_name in file_names:
    # Construct the full URL
    full_url = base_url + file_name
    
    # Download the file
    response = requests.get(full_url, stream=True)
    if response.status_code == 200:
        # Save the file to the directory
        file_path = os.path.join(download_directory, file_name)
        with open(file_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=8192):
                file.write(chunk)
        
        # Unzip the file to the specified location
        with zipfile.ZipFile(file_path, 'r') as zip_ref:
            zip_ref.extractall(unzip_directory)
        
        print(f"Downloaded and extracted {file_name}")
    else:
        print(f"Failed to download {file_name}")

print("All files processed.")
```
